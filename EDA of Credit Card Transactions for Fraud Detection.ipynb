# Project: EDA of Credit Card Transactions for Fraud Detection

## Problem Statement:
Credit card fraud is a significant concern for financial institutions and
consumers. Fraudulent activities cause substantial losses and impact
customer trust. This project aims to perform Exploratory Data Analysis
(EDA) on credit card transaction data to identify patterns and anomalies
that may indicate fraudulent behavior.

Dataset source: Kaggle 
https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
# To find the column names and their datatypes
df.info()
```

```python
# To find the missing values
df.isnull().sum()
```

```python
# Defining Fraud and Legitimate classes
fraud = df["Class"]==1
legit = df["Class"]==0
```

```python
# Sum of Legitimate and Fraud transactions

fraud_count =  sum(fraud)
print ("Number of Fraudulent transactions:",fraud_count)
legit_count =  sum(legit)
print ("Number of Safe transactions:",legit_count)
total = legit_count+fraud_count
print ("Total number of transactions:",total)
```

```python
# Percentage of Fraudulent transactions
fraud_percent = (fraud_count/total)*100
print (fraud_percent,"%")
```

```python
df.describe()
```

```python
# minimum, maximum, mean, and median values for column "Amount"
print ("Minimum:",df["Amount"].min())
print ("Maximum:",df["Amount"].max())
print ("Mean:",df["Amount"].mean())
print ("Median:",df["Amount"].median())
```

```python
# Bar chart for showing count of fraudulent and legitimate transactions

categories = ['Legitimate', 'Fraudulent']
counts = [legit_count, fraud_count]

# Plot
plt.bar(categories, counts, color=['green', 'red'], width=0.5)
plt.title('Transaction Counts: Legitimate vs Fraudulent')
plt.ylabel('Count')
plt.xlabel('Transaction Type')
plt.show()
```

```python
# Plotting the bar chart with a logarithmic scale because the bar showing count of fraudulent transactions is not much 
# visible due to magnitude difference of values. This change of scale won't affect the relation between the two counts.

plt.bar(categories, counts, color=['green', 'red'], width=0.5)  
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.title('Transaction Counts: Legitimate vs Fraudulent (Log Scale)')
plt.ylabel('Count (Log Scale)')
plt.xlabel('Transaction Type')
plt.show()
```

```python
# Histogram of transaction amounts
plt.hist(df["Amount"], bins=70, color='skyblue', edgecolor='black')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Transaction Amounts')
plt.show()
```

```python
# Heatmap to visualize the correlation between numerical features
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, linewidths=0.5, square=True, cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
